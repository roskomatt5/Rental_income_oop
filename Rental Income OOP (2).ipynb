```python
class rentalIncome:
    def __init__(self):
        self.monthly_income = 0
        self.monthly_expenses = 0
        self.total_investment = 0
        self.mcf = self.monthly_income - self.monthly_expenses
    def incomeAdd(self):
        self.monthly_income_input = []
        self.income_category = []
        rental_income = float(input("Please provide the amount of Rental Income you plan on receiving on a monthly basis "))
        self.income_category.append('Rental Income')  
        self.monthly_income_input.append(rental_income)
        laundry = float(input("Please provide the amount of income you plan on receiving from laundry machines provided on a monthly basis "))
        self.income_category.append('Laundry')  
        self.monthly_income_input.append(laundry)  
        storage = float(input("Please provide the amount of income you plan on receiving for tennant's storage on a monthly basis "))
        self.income_category.append('Storage')  
        self.monthly_income_input.append(storage)
        misc = float(input("Please provide any other amount for miscellanous sources of monthly income. "))
        self.income_category.append('misc')  
        self.monthly_income_input.append(misc)
        for i in range(len(self.income_category)):
            print(f"{self.income_category[i]}: ${self.monthly_income_input[i]}")
        self.monthly_income = sum(self.monthly_income_input)
        print(f"your total monthly income is ${self.monthly_income}!") 

#       self.monthly_income_input = []
#         self.income_category = []
#         num_inputs = int(input("How many sources of income does the property have? "))   
#         for i in range(num_inputs):  
#             user_category = input(f'Pease enter income source #{i + 1}: ').lower()
#             self.income_category.append(user_category)
#             user_amount = float(input('Pease enter an estimation of monthly income from the category provided: '))
#             self.monthly_income_input.append(user_amount)
#         for i in range(len(self.income_category)):
#             print(f"{self.income_category[i]}: ${self.monthly_income_input[i]}")
#         self.monthly_income = sum(self.monthly_income_input)
#         print(f"your total monthly income is ${self.monthly_income}!") 
        
    def expensesAdd(self):
        self.monthly_expense_input = []
        self.expenses_category = []
        if self.monthly_income == 0:
          print("In order for us to properly calculate vacancy costs please finish calculating income.")
        else:
            taxes = float(input("Please provide an estimate for the monthly cost for taxes. "))
            self.expenses_category.append('Taxes')  
            self.monthly_expense_input.append(taxes)
            insurance = float(input("Please provide an estimate for the monthly cost for insurance. "))
            self.expenses_category.append('Insurance')  
            self.monthly_expense_input.append(insurance)
####utilities
            electric = float(input("Please provide an estimate for the monthly cost for Electiric Utility Costs. "))
            self.expenses_category.append('Utilities: Electric')  
            self.monthly_expense_input.append(electric)
            water = float(input("Please provide an estimate for the monthly cost for Water Utility costs. "))
            self.expenses_category.append('Utilities: Water')  
            self.monthly_expense_input.append(water)
            sewage = float(input("Please provide an estimate for the monthly cost for Sewage Utility Costs. "))
            self.expenses_category.append('Utilities: Sewage')  
            self.monthly_expense_input.append(sewage)
            gas = float(input("Please provide an estimate for the monthly cost for Gas Utility costs. "))
            self.expenses_category.append('Utilities: Gas')  
####utilities
            self.monthly_expense_input.append(gas)
            capex = float(input("Please provide an estimate for the monthly cost for CapEx costs. "))
            self.expenses_category.append('CapEx')  
            self.monthly_expense_input.append(capex)
            mgt = float(input("Please provide an estimate for the monthly cost for on site Management costs. "))
            self.expenses_category.append('Management Costs')  
            self.monthly_expense_input.append(mgt)
            mortgage = float(input("Please provide an estimate for the monthly cost for Mortgage costs. "))
            self.expenses_category.append('Mortgage')  
            self.monthly_expense_input.append(mortgage)

            for i in range(len(self.expenses_category)):
                print(f"{self.expenses_category[i]}: ${self.monthly_expense_input[i]}")
                self.monthly_expenses = sum(self.monthly_expense_input)
            print(f"your total monthly expenses are ${self.monthly_expenses}!")
#         self.monthly_expense_input = []
#         self.expenses_category = []
#         num_inputs = int(input("How many sources of expenses does the property have? "))
#         if self.monthly_income == 0:
#             print("In order for us to properly calculate vacancy costs please finish calculating income.")
#         else:
#             for i in range(num_inputs):  
#                 user_category = input(f'Please enter expenses source #{i + 1}: ').lower()
#                 self.expenses_category.append(user_category)
#                 user_amount = float(input('Pease enter an estimation of expenses from the category provided: '))
#                 self.monthly_expense_input.append(user_amount)
#             for i in range(len(self.expenses_category)):
#                 print(f"{self.expenses_category[i]}: ${self.monthly_expense_input[i]}")
#             vacancy_costs = self.monthly_income * .05
#             self.monthly_expenses = (sum(self.monthly_expense_input) + vacancy_costs)
#         print(f"your monthly expenses including vacancy costs are ${self.monthly_expenses}")

   
    def getTotalInvestment(self):
        investments = []
        if self.monthly_income == 0:
           print('please finsish calculating monthly income')
        if self.monthly_expenses == 0:
           print('please finsish calculating monthly expenses')
        else:
            down_payment = float(input('What was the amount of your down payment? '))
            investments.append(down_payment)
            rehab_budget = float(input('Do you have any rehab costs? If so, please enter the amount it will cost.If not type "0" '))
            investments.append(rehab_budget)
            misc_other = float(input('Are ther any other miscelanous costs? If so, please enter the sum of those costs. If not type "0".' ))    
            investments.append(misc_other)
                                 
            self.total_investment = sum(investments)
            cashFlow = self.monthly_income - self.monthly_expenses
            print(f"Your total return on investment is {float(((cashFlow * 12) / self.total_investment) * 100)}%")
        

        
run_calculator = rentalIncome()

def run():
    while True:
        response = input('What do you need to calculate? income / expenses / roi? ')

        if response.lower() == 'income':
            run_calculator.incomeAdd()
            
        if response.lower() == 'expenses':
            run_calculator.expensesAdd()
            
        if response.lower() == 'roi':
            run_calculator.getTotalInvestment()
    
run()       





        
    
```
